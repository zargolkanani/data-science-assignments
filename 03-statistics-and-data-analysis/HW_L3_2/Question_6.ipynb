#### Question 6: Replacing Missing Values Using the Mean
Description: Create a DataFrame from the following list of dictionaries, and then fill the missing values (NaNs) with the mean of each respective column.

data = [{'a': 1, 'b': np.nan}, {'a': 3, 'b': 4}, {'a': np.nan, 'b': 2}]


```python
import pandas as pd
import numpy as np
```

```python
data = [
    {'a': 1, 'b': np.nan},
    {'a': 3, 'b': 4},
    {'a': np.nan, 'b': 2}
]

df = pd.DataFrame(data) 
print(df)
```

```python
df = df.fillna(df.mean())  # Fill NaN values with the mean of each column
df
```
