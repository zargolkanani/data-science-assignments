#### Question 1: Creating a New Column and Filtering a Pandas DataFrame
Description: Using the provided Pandas DataFrame containing the area and population of US states, calculate the population density for each state and add it as a new column to this DataFrame. Then, filter and extract the states with a density greater than 100.

```python
import pandas as pd
```

```python
states = pd.DataFrame({
    "area": [423967, 695662, 141297, 170312, 149995],
    "pop": [38332521, 26448193, 19651127, 19552860, 12882135]
}, index=["CA", "TX", "NY", "FL", "IL"])

states
```

```python
states["density"] = states["pop"] / states["area"]
states
```

```python
filtered_states = states[states["density"] > 100] # Filter states with density greater than 100
filtered_states
```
