#### Question 2: Creating a DataFrame and Modifying Arrays Based on a Filter
Description: First, create a $5 \times 5$ DataFrame containing random integers. Then, replace any values that are higher than the overall mean with the string "High".

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(1, 100, size=(5, 5)), columns=list("ABCDE"))  # Create a 5x5 DataFrame with random integers
df
```

```python
mean = df.values.mean()
mean
```

```python
result = df.mask(df > mean, other = "High")
print(result)
```
