#### Question 4: Shuffling the Rows of a DataFrame
Description: Randomly shuffle the rows of the following DataFrame. (Note: The dimensions have been corrected to $3 \times 3$ to match the 3 specified columns).
df = pd.DataFrame(np.random.rand(3, 10), columns=["A", "B", "C"])

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.rand(10, 3), columns=["A", "B", "C"])
df
```

```python
shuffeled_df = df.sample(frac=1).reset_index(drop=True)  # Shuffle the DataFrame rows
shuffeled_df
```
