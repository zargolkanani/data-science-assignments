#### Question 5: Using loc and iloc for Element Selection
Description: Consider the following Series. Select the second and third elements of this Series using both loc and iloc.

s = pd.Series([40, 30, 20, 10], index=['w', 'x', 'y', 'z'])


```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([40, 30, 20, 10], index=['w', 'x', 'y', 'z'])
s
```

```python
print(s.iloc[1:3])  # Access elements at index positions 1 and 2
```

```python
print(s.loc['x':'y'])  # Access elements from index 'x' to 'y' (inclusive)
```
