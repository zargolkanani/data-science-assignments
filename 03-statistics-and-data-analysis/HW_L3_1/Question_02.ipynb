#### Question 2: Extracting Common Elements of Two Arrays
Create two 1D NumPy arrays, A and B, each containing 10 random integers from 1 to 20. Use seed 14 so the random results can be repeated. Then, extract the unique common elements between these two arrays.

```python
import numpy as np
```

```python
np.random.seed(14)  
array_A = np.random.randint(1, 21, size=(10)) # Generate an array of 10 random integers between 1 and 20
array_B = np.random.randint(1, 21, size=(10)) # Generate an array of 10 random integers between 1 and 20
print("Array A:", array_A)
print("Array B:", array_B)
```

```python
common_elements = np.intersect1d(array_A, array_B) # Find common elements
common_elements
```

```python
unique_elements = np.setxor1d(array_A, array_B)# Find unique elements
unique_elements
```
