#### Question 6: Selecting Numbers Using a Mask
Create a 1D array of 10 random integers between 0 and 50. Use boolean masking to select the elements that are divisible by 3 but not divisible by 6.

```python
import numpy as np
```

```python
arr = np.random.randint(0, 50, size=(10)) # Generate an array of 10 random integers between 0 and 50
print ("Array:", arr)
```

```python
mask = (arr % 3 == 0) & (arr % 6 != 0) # Filter the array to include only elements that are divisible by 3 but not by 6
arr = arr[mask] # Apply the mask to filter the array
print ("Filtered Array:", arr)
```
