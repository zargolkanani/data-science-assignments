#### Question 3: Non-Negative Random Float Numbers
Create a $5 \times 5$ array of random float numbers between -1 and 1, and then replace all the negative values with zero. Use seed 14 for reproducibility.

```python
import numpy as np
```

```python
np.random.seed(14)  # Set a random seed for reproducibility
arr = np.random.uniform(-1, 1, size=(5, 5))  # Generate a 5x5 array of random numbers uniformly distributed between -1 and 1
arr
```

```python
result = np.where(arr < 0, 0, arr)  # Replace negative values with 0
result
```
