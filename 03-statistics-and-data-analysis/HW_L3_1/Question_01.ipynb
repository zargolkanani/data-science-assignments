#### Question 1: Reshaping and Extracting NumPy Sub-arrays
Create a 1D NumPy array with 1000 elements containing integers from 0 to 1000. Then, reshape it into a $20 \times 50$ array. After that, create another array whose elements are the even rows and odd columns of the first array. 

```python
import numpy as np
```

```python
arr = np.arange(1000) # Create an array of integers from 0 to 999
arr
```

```python
reshaped_arr = arr.reshape(20, 50)  # Reshape the array to 20 rows and 50 columns
reshaped_arr
```

```python
result = reshaped_arr[::2, 1::2]  # Slice the array
result
```
