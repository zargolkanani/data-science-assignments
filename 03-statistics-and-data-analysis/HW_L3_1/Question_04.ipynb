#### Question 4: Calculating Mean, Median, and Standard Deviation
Create an array of 1000 elements with a normal distribution, a mean of 50, a standard deviation of 5, and seed 14. Then, calculate the mean, median, and standard deviation of the numbers inside this array.

```python
import numpy as np
```

```python
np.random.seed(14) 
arr = np.random.normal(loc=50, scale=5, size=(1000))
arr
```

```python
print ("Mean:", np.mean(arr))
print ("Median:", np.median(arr))
print ("Standard Deviation:", np.std(arr))
```
