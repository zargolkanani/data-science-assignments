#### Question 5: Swapping Axes of a 3D Array
Create a 3D array containing random integers between 0 and 50 with dimensions $(3, 4, 5)$. Then, swap (interchange) the first axis and the third axis of this array.

```python
import numpy as np
```

```python
arr = np.random.randint(0, 50, size=(3, 4, 5))  # Generate a 3x4x5 array of random integers between 0 and 
arr
```

```python
resualt = np.swapaxes(arr, 0, 2)  # Swap the first and second axes of the array
resualt
```

```python
print("Original array shape:", arr.shape)
print("Array shape after swapping axes:", resualt.shape)
```
